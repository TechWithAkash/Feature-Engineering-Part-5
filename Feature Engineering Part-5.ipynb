Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
you have collected data on the amount of time students spend studying for an exam and their final exam
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

=>
To calculate the Pearson correlation coefficient between two variables, you need to follow these steps:

1. Calculate the mean (average) of each variable.
2. Calculate the difference between each data point and its respective mean for both variables.
3. Multiply the differences for each data point and sum up these products.
4. Calculate the standard deviation of each variable.
5. Divide the sum of products from step 3 by the product of the standard deviations from step 4.

The formula for the Pearson correlation coefficient (r) is:

\[ r = \frac{\sum{(X_i - \bar{X})(Y_i - \bar{Y})}}{\sqrt{\sum{(X_i - \bar{X})^2} \sum{(Y_i - \bar{Y})^2}}} \]

Where:
- \( X_i \) and \( Y_i \) are individual data points for the two variables.
- \( \bar{X} \) and \( \bar{Y} \) are the means of the two variables.

Let's assume you have collected the following data:

Time Spent Studying (hours):
\[ [10, 5, 15, 20, 8, 12, 4, 18, 6, 14] \]

Final Exam Scores:
\[ [90, 75, 95, 100, 80, 92, 70, 98, 78, 93] \]

Step 1: Calculate the means of both variables.
\[ \bar{X} = \frac{10 + 5 + 15 + 20 + 8 + 12 + 4 + 18 + 6 + 14}{10} = 11.2 \]

\[ \bar{Y} = \frac{90 + 75 + 95 + 100 + 80 + 92 + 70 + 98 + 78 + 93}{10} = 87.1 \]

Step 2: Calculate the differences between data points and their means for both variables.

For Time Spent Studying:
\[ (X_i - \bar{X}) \]
\[ (10 - 11.2), (5 - 11.2), (15 - 11.2), (20 - 11.2), (8 - 11.2), (12 - 11.2), (4 - 11.2), (18 - 11.2), (6 - 11.2), (14 - 11.2) \]

For Final Exam Scores:
\[ (Y_i - \bar{Y}) \]
\[ (90 - 87.1), (75 - 87.1), (95 - 87.1), (100 - 87.1), (80 - 87.1), (92 - 87.1), (70 - 87.1), (98 - 87.1), (78 - 87.1), (93 - 87.1) \]

Step 3: Multiply the differences for each data point and sum up these products.

\[ \sum{(X_i - \bar{X})(Y_i - \bar{Y})} = (-2.12 * 2.9) + (-6.12 * -12.1) + (3.88 * 7.9) + (8.88 * 12.9) + (-3.12 * -0.1) + (0.88 * 4.9) + (-7.12 * -17.1) + (6.88 * 10.9) + (-5.12 * -9.1) + (2.88 * 5.9) \]

This sum is the numerator of the Pearson correlation coefficient.

Step 4: Calculate the standard deviation of each variable.

For Time Spent Studying:
\[ \sqrt{\sum{(X_i - \bar{X})^2}} = \sqrt{(-1.12)^2 + (-6.12)^2 + (3.88)^2 + (8.88)^2 + (-3.12)^2 + (0.88)^2 + (-7.12)^2 + (6.88)^2 + (-5.12)^2 + (2.88)^2} \]

For Final Exam Scores:
\[ \sqrt{\sum{(Y_i - \bar{Y})^2}} = \sqrt{(2.9)^2 + (-12.1)^2 + (7.9)^2 + (12.9)^2 + (-0.1)^2 + (4.9)^2 + (-17.1)^2 + (10.9)^2 + (-9.1)^2 + (5.9)^2} \]

Step 5: Divide the sum of products from step 3 by the product of the standard deviations from step 4.

Now, plug these values into the formula for the Pearson correlation coefficient (r):

\[ r = \frac{\sum{(X_i - \bar{X})(Y_i - \bar{Y})}}{\sqrt{\sum{(X_i - \bar{X})^2} \sum{(Y_i - \bar{Y})^2}}} \]

After performing the calculations, you'll get the Pearson correlation coefficient (r), which will be a value between -1 and 1.

Interpretation:
- If r is close to 1, it indicates a strong positive linear relationship between the time spent studying and final exam scores. In other words, as students spend more time studying, their exam scores tend to increase.
- If r is close to -1, it indicates a strong negative linear relationship, suggesting that more study time is associated with lower exam scores.
- If r is close to 0, it suggests little to no linear relationship between the two variables.

The interpretation depends on the actual value of the calculated Pearson correlation coefficient.

Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables.
Suppose you have collected data on the amount of sleep individuals get each night and their overall job
satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two
variables and interpret the result.

=>
Spearman's rank correlation coefficient, often denoted as ρ (rho), measures the strength and direction of the monotonic relationship between two variables. Unlike Pearson's correlation, Spearman's correlation does not assume a linear relationship and is more appropriate when dealing with ordinal or non-normally distributed data. It is calculated based on the ranks of the data points rather than their actual values.

Here are the steps to calculate Spearman's rank correlation:

1. Rank the data for each variable separately.
2. Calculate the difference between the ranks for each pair of data points.
3. Square these differences.
4. Sum up the squared differences.
5. Use the formula to calculate Spearman's rank correlation coefficient.

Let's assume you have collected the following data:

Amount of Sleep (hours):
\[ [6, 7, 5, 8, 5, 6, 7, 8, 6, 9] \]

Job Satisfaction Level (on a scale of 1 to 10):
\[ [8, 6, 5, 9, 4, 7, 8, 9, 6, 9] \]

Step 1: Rank the data for each variable separately.

For Amount of Sleep:
\[ [5, 6, 5, 7, 5, 6, 6, 8, 7, 9] \]

For Job Satisfaction Level:
\[ [7, 4, 3, 8, 2, 6, 7, 8, 4, 8] \]

Step 2: Calculate the difference between the ranks for each pair of data points.

\[ [0, 0, 2, 0, 3, 0, 1, 1, 3, 1] \]

Step 3: Square these differences.

\[ [0, 0, 4, 0, 9, 0, 1, 1, 9, 1] \]

Step 4: Sum up the squared differences.

Sum of squared differences = 25

Step 5: Use the formula to calculate Spearman's rank correlation coefficient (ρ):

\[ \rho = 1 - \frac{6 \sum(\text{Squared Rank Differences})}{n(n^2 - 1)} \]

Where:
- \( \sum(\text{Squared Rank Differences}) \) is the sum of squared rank differences (which we calculated as 25).
- \( n \) is the number of data points (which is 10 in this case).

Now, plug these values into the formula:

\[ \rho = 1 - \frac{6 \times 25}{10(10^2 - 1)} \]

\[ \rho = 1 - \frac{150}{990} \]

\[ \rho = 1 - 0.1515 \]

\[ \rho \approx 0.8485 \]

Interpretation:
Spearman's rank correlation coefficient (ρ) is approximately 0.8485. This value suggests a strong positive monotonic relationship between the amount of sleep individuals get each night and their overall job satisfaction level. In other words, as the amount of sleep increases, job satisfaction tends to increase as well, and as the amount of sleep decreases, job satisfaction tends to decrease. The strength of this relationship is relatively high, indicating that sleep and job satisfaction are closely related in a monotonic fashion.
